```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
#Set parameters
nx = 41
L = 2.0
dx = L/ (nx-1)
nu = 0.3 #Viscosity
sigma = 0.2 #new sigma limit to be considered
dt = sigma * dx**2 / nu #Sigma is important in time step definition
nt = 20

#Get the grid point coordinates
x = numpy.linspace(0.0, L, num=nx)

#Set initial conditions
u0 = numpy.ones(nx)
mask = numpy.where(numpy.logical_and(x>=0.5, x<=1.0))
u0[mask] = 2.0
```

```python
#Integrate in time 
u = u0.copy()
for n in range(nt):
    u[1:-1] = u[1:-1] + nu * dt / dx**2 * (u[2:] - 2 * u[1:-1] + u[:-2])
```

```python
pyplot.figure(figsize=(6.0,4.0))
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, label='Initial', color='C1', linestyle='--')
pyplot.plot(x, u, label = 'nt = {}'.format(nt), color = 'C2', linestyle='-')
pyplot.legend(loc='upper right')
pyplot.xlim(0.0, L)
pyplot.ylim(0.5, 2.5);
```

```python
#Animations
from matplotlib import animation
from IPython.display import HTML
```

```python
def diffusion(u0, sigma=0.5, nt=20):
    u_hist = [u0.copy()]
    u = u0.copy()
    for n in range(nt):
        u[1:-1] = u[1:-1] + sigma * (u[2:] - 2 * u[1:-1] + u[:-2])
        u_hist.append(u.copy())
    return u_hist
```

```python
#compute history of the numerical solution
u_hist = diffusion(u0, sigma=sigma, nt=nt)
```

```python
fig = pyplot.figure(figsize=(6.0,4.0))
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
line = pyplot.plot(x, u0, color='C1', linestyle='-', linewidth=2)[0]
pyplot.xlim(0.0, L)
pyplot.ylim(0.5, 2.5)
fig.tight_layout()
```

```python
def update_plot(n, u_hist):
    fig.suptitle('Time step {:0>2}'.format(n))
    line.set_ydata(u_hist[n])
```

```python
#Creating the animation 
anim = animation.FuncAnimation(fig, update_plot, frames=nt, fargs=(u_hist,),interval=100)
```

```python
#Display the video
HTML(anim.to_html5_video())
```
