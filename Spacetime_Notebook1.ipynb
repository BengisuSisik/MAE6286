```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
#Set parameters for 1D linear convection equation
nx = 41
L = 2.0
dx = L / (nx-1)
nt = 25
dt = 0.02
c = 1.0

#Definition of grid point coordinates
x = numpy.linspace(0.0, L, num =nx)
```

```python
#First, set initial conditions with 1.0 

u0 = numpy.ones(nx)

#Get a list of indices where 0.5 <= x <= 1.0
mask = numpy.where(numpy.logical_and(x >= 0.5, x<=1.0))
print(mask)
```

```python
u0[mask] = 2.0
print(u0)
```

```python
#Plot the initial conditions
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('Initial Conditions')
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0,linestyle='--')
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```

```python
u = u0.copy()
for n in range(1, nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - c*dt/dx * (un[i]-un[i-1])
    
```

```python
#Plot the solution after nt time steps together with the initial conditions
pyplot.figure(figsize=(4.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, label = 'Initial', linestyle = '--', color = 'C1')
pyplot.plot(x, u, label = 'nt = 25', linestyle = '-', color = 'C2')
pyplot.legend()
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```

```python
#Non-linear convection
nx = 41
L = 2.0
dx = L / (nx-1)
nt = 15
dt = 0.02

#This time not c

#Definition of grid point coordinates
x = numpy.linspace(0.0, L, num =nx)
u0 = numpy.ones(nx)
mask = numpy.where(numpy.logical_and(x >= 0.5, x<=1.0))
u0[mask] = 2.0
```

```python
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('Initial conditions')
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, color='C0', linestyle='--', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```

```python
u = u0.copy()
for n in range(1, nt):
    u[1 :] = u[1:] -dt /dx * u[1:] * (u[1:]-u[:-1])
```

```python
pyplot.figure(figsize=(4.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, label='Initial',
            color='C0', linestyle='--', linewidth=2)
pyplot.plot(x, u, label='nt = {}'.format(nt),
            color='C1', linestyle='-', linewidth=2)
pyplot.legend()
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```
